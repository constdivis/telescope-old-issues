```python
issues = data['ar_magazins'].set_index('id').to_dict(orient = 'index')
issues[173]
```

```python
for k, a in all_articles.items():
    a_id = a['id']
    for _, m in rel_magazins_articles.items():
        if a_id == m['fk_article']:
            a['year'] = issues[m['fk_magazin']]['year']
            a['number'] = issues[m['fk_magazin']]['number']
```

```python
authors = data['ar_authors'].set_index('id').to_dict(orient = 'index')
authors[278]
```

```python
for k, v in authors.items():
    person = re.findall(r'^(\w{1}\.)\s(\w+)\Z', v['author'])
    if person:
        v['person'] = f'{person[0][1]} {person[0][0]}'
    else:
        v['person'] = ''
```
